```python
# Import modules

import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os
```

```python
# Open image

img = mpimg.imread("exemple.jpg")

plt.imshow(img)
plt.show()
```

```python
# Select pixels

corners = [
    (160, 127),
    (119, 800),
    (453, 800),
    (479, 127)
]

circles_img = img.copy()
for corner in corners:
    circles_img = cv2.circle(circles_img, corner, 20, (0, 255, 0), -1)

plt.imshow(circles_img)
plt.show()
```

```python
# Using corners to remove perspective

h, w = img.shape[:2]

src = np.float32(corners)

dst = [
    [0, h], 
    [0, 0], 
    [w, 0], 
    [w, h]
]

dst = np.float32(dst)

transformation_matrix = cv2.getPerspectiveTransform(src, dst)
warped_img = cv2.warpPerspective(img, transformation_matrix, (w, h), flags=cv2.INTER_LINEAR)

plt.imshow(warped_img)
plt.show()
```
